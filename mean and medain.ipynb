```python
arr = np.array([50, 50, 67, 130, 122, 143, 134, 144, 80, 312, 212, 160, 170, 165, 2000])
arr.sort()
print(arr)
mm = np.median(arr)
print(mm)

me = np.mean(arr)
print(me)
```

```python
plt.hist(arr, bins=40, edgecolor='black')
plt.axvline(mm, color='red', linestyle='--', linewidth=2, label=f'Median: {mm}')
plt.axvline(me, color='yellow', linestyle='--', linewidth=2, label=f'Mean: {me}')
```
